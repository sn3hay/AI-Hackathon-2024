```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import json
import numpy as np

software_occupation_ids = ["101005", "101034", "101026", "101020", "101010"]
tech_occupation_ids = ["151256", "151212", "151257", "151211", "151244"]

url = "https://145dcd62-2d7e-409c-9c15-0bcc31fc9fda.mock.pstmn.io"

def fetch_occupation_data(url, occupation_id, region_id):
    url = "{}/projected-occupation-growth/dimensions/lotocc".format(url)

    payload = json.dumps({
        "id": occupation_id,
        "region": {
            "level": "county",
            "id": region_id
        }
    })
    headers = {"Content-Type": "application/json"}
    response = requests.get(url, headers=headers, data=payload)
    return response.json()['data']

def plot_occupation_data():    
    region_id = ["29", "15", "37", "3", "13"]  # Assuming "5" is the ID for Erie County

    timeframes = ["oneYear", "threeYear", "fiveYear", "tenYear"]

    plt.figure(figsize=(10, 6))

    for reg_id, occ_id in zip(region_id, tech_occupation_ids):
        data = fetch_occupation_data(url, occ_id, reg_id)

        df = pd.DataFrame([data])

        for timeframe in timeframes:
            df[f"{timeframe}Projected"] = df["baseEmployment"] * (1 + df["growthPercent"].apply(lambda x: x[timeframe]))

        base_employment = df["baseEmployment"].values[0]
        projected_employment = [df[f"{timeframe}Projected"].values[0] for timeframe in timeframes]

        x_labels = ["Base", "1 Year", "3 Years", "5 Years", "10 Years"]
        x_values = range(len(x_labels))

        plt.plot(x_values, [base_employment] + projected_employment, marker='o', linestyle='-', label=df["name"].values[0])

        plt.xlabel("Timeframe")
        plt.ylabel("Number of Jobs")
        plt.title("Projected Growth of Software Jobs in Erie County")
        plt.xticks(x_values, x_labels)
        plt.grid(True)
        plt.legend()

        plt.tight_layout()
        plt.show()

def plot_feeder_job(dimension, occupation_id):
    url = "{}/career-pathways/dimensions/{}/feederjobs".format(url, dimension)
    payload = json.dumps({
                "id": occupation_id, 
                "categories": [
                    "Advancement",
                    "LateralAdvancement",
                    "Similar",
                    "LateralTransition"
                ],
                "region": {
                    "nation": "us"
                },
                "limit": 3
            })
    headers = {"Content-Type": "application/json"}
    response = requests.get(url, headers=headers, data=payload)
    response_data = response.json()

    cto_salary = response_data["data"]["meanSalary"]
    pathways = response_data["data"]["pathways"]
    pathway_names = [p['name'] for p in pathways]
    pathway_salaries = [p['meanSalary'] for p in pathways]
    salary_diffs = [p['meanSalaryDiff'] for p in pathways]

    plt.figure(figsize=(10, 6))

    plt.bar(pathway_names, pathway_salaries, color='skyblue', label="Mean Salary")

    plt.axhline(cto_salary, color='red', linestyle='--', label="CTO Salary")

    plt.xlabel('Job Titles')
    plt.ylabel('Salary ($)')
    plt.title('Salary Comparison with Chief Technology Officer')

    for i, diff in enumerate(salary_diffs):
        plt.text(i, pathway_salaries[i] + 2000, f"${diff}", ha='center', va='bottom', fontsize=9)

    plt.legend()

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()

    plt.show()

def plot_nextstep_jobs(dimension, occupation_id):
    url = "{}/career-pathways/dimensions/{}/nextstepjobs".format(url, dimension)
    payload = json.dumps({
                "id": occupation_id, 
                "categories": [
                    "Advancement",
                    "LateralAdvancement",
                    "Similar",
                    "LateralTransition"
                ],
                "region": {
                    "nation": "us"
                },
                "limit": 3
            })
    headers = {"Content-Type": "application/json"}
    response = requests.get(url, headers=headers, data=payload)
    response_data = response.json()

    web_dev_salary = response_data["data"]["meanSalary"]
    web_dev_name = response_data["data"]["name"]
    pathways = response_data["data"]["pathways"]

    pathway_names = [p['name'] for p in pathways]
    pathway_salaries = [p['meanSalary'] for p in pathways]
    salary_diffs = [p['meanSalaryDiff'] for p in pathways]

    pathway_names.insert(0, web_dev_name)
    pathway_salaries.insert(0, web_dev_salary)
    salary_diffs.insert(0, 0)  # Web Developer has no salary difference from itself

    plt.figure(figsize=(10, 6))

    plt.bar(pathway_names, pathway_salaries, color='lightblue', edgecolor='black')

    for i, diff in enumerate(salary_diffs):
        plt.text(i, pathway_salaries[i] + 2000, f"${diff}", ha='center', va='bottom', fontsize=9)

    plt.title("Salary Comparison for Web Developer Career Pathways", fontsize=14)
    plt.xlabel("Job Titles")
    plt.ylabel("Salary ($)")
    plt.xticks(rotation=45, ha='right')

    plt.axhline(web_dev_salary, color='red', linestyle='--', label=f"{web_dev_name} Salary (${web_dev_salary})")

    plt.legend()

    plt.tight_layout()

    plt.show()

def plot_broad_comp_data():
    estimate_comp_url = "{}/comp-emsicloud/estimate".format(url)
    payload = json.dumps({
        "titles": ["ETCC677E49FBA73537"], # Java web developer title
        "msa": "15380" # Buffalo-Niagara-Cheektowaga MSA code
    })
    headers = {"Content-Type": "application/json"}
    response = requests.get(estimate_comp_url, headers=headers, data=payload)
    data = response.json()

    national_percentiles = data['national_estimate']['percentiles']
    regional_percentiles = data['regional_estimate']['percentiles']

    national_salaries = [item['annual_salary'] for item in national_percentiles]
    regional_salaries = [item['annual_salary'] for item in regional_percentiles]
    percentiles = [item['percentile'] for item in national_percentiles]

    plt.figure(figsize=(10, 6))
    bar_width = 0.35
    index = np.arange(len(percentiles))

    plt.bar(index, national_salaries, bar_width, label='National', color='skyblue', edgecolor='black')
    plt.bar(index + bar_width, regional_salaries, bar_width, label='Regional (Buffalo-Cheektowaga-Niagara Falls, NY)', color='lightgreen', edgecolor='black')

    plt.xlabel('Percentiles')
    plt.ylabel('Annual Salary ($)')
    plt.title('National vs Regional Salary Estimates by Percentile for Java Web Developers')
    plt.xticks(index + bar_width / 2, percentiles)
    plt.legend()

    for i in range(len(national_salaries)):
        plt.text(i, national_salaries[i] + 2000, f"${national_salaries[i]:,}", ha='center')
        plt.text(i + bar_width, regional_salaries[i] + 2000, f"${regional_salaries[i]:,}", ha='center')

    plt.tight_layout()
    plt.show()

    salary_diff = [regional_salaries[i] - national_salaries[i] for i in range(len(national_salaries))]

    plt.figure(figsize=(10, 6))
    plt.bar(percentiles, salary_diff, color='lightcoral', edgecolor='black')
    plt.title('Difference in Salary: Regional vs National by Percentile for Java Web Developers')
    plt.xlabel('Percentiles')
    plt.ylabel('Salary Difference ($)')
    plt.axhline(0, color='black', linewidth=1)

    for i, diff in enumerate(salary_diff):
        plt.text(percentiles[i], diff + 1000 if diff > 0 else diff - 3000, f"${diff:,}", ha='left')

    plt.tight_layout()
    plt.show()

def plot_title_comp_data():
    estimate_comp_url = "{}/role-pricing/comp-emsicloud/estimate".format(url)
    payload_list = [
            {
                "country_code": "US",
                "title_name": "Data Scientist",
                "min_years_experience": 10
            },
            {
                "country_code": "US",
                "title_name": "Programmer",
                "min_years_experience": 10,
                "edu_level": 3
            },
            {
                "country_code": "US",
                "title_name": "Web Developer",
                "min_years_experience": 7,
                "edu_level": 2,
                "skills_name": [
                "javascript",
                "react"
                ]
            },
            {
                "country_code": "US",
                "title_name": "Computer Systems Analysts",
                "min_years_experience": 3,
                "edu_level": 1
            },
            {
                "country_code": "US",
                "title_name": "Information Security Analysts",
                "min_years_experience": 5,
                "edu_level": 5,
                "skills_name": [
                "comptia+",
                "linux"
                ]
            }
    ]
    payload = json.dumps(payload_list)
    headers = {"Content-Type": "application/json"}
    response = requests.post(estimate_comp_url, headers=headers, data=payload)
    response_data = response.json()['data']
    min_salaries = [job["min_salary"] for job in response_data]
    max_salaries = [job["max_salary"] for job in response_data]
    avg_salaries = [job["average_salary"] for job in response_data]

    plt.figure(figsize=(12, 7))

    job_titles = [x["title_name"] for x in payload_list]
    # Plotting min, average, and max salaries as line plots
    plt.plot(job_titles, min_salaries, marker='o', label='Min Salary', color='lightblue', linestyle='-')
    plt.plot(job_titles, avg_salaries, marker='o', label='Average Salary', color='lightgreen', linestyle='--')
    plt.plot(job_titles, max_salaries, marker='o', label='Max Salary', color='lightcoral', linestyle='-.')

    # Add labels and title
    plt.xlabel('Job Titles')
    plt.ylabel('Salary ($)')
    plt.title('Salary Range for Different Job Titles (Min, Max, and Average)')
    plt.xticks(rotation=45, ha='right')
    plt.legend()

    # Annotate points with salary values
    for i, title in enumerate(job_titles):
        # plt.text(i, min_salaries[i] - 5000, f"${min_salaries[i]:,.2f}", ha='center', va='top')
        # plt.text(i, avg_salaries[i] + 2000, f"${avg_salaries[i]:,.2f}", ha='center', va='bottom')
        # plt.text(i, max_salaries[i] + 5000, f"${max_salaries[i]:,.2f}", ha='center', va='bottom')

        plt.tight_layout()
        plt.show()

def plot_combined_title_occupation_data():
    region_id = ["29", "15", "37", "3", "13"]
    timeframes = ["oneYear", "threeYear", "fiveYear", "tenYear"]
    
    for reg_id, occ_id in zip(region_id, tech_occupation_ids):
        # Fetching occupation data for each region and occupation
        data = fetch_occupation_data(url, occ_id, reg_id)
        df = pd.DataFrame([data])
    
        # Calculate projected employment for each timeframe
        for timeframe in timeframes:
            df[f"{timeframe}Projected"] = df["baseEmployment"] * (1 + df["growthPercent"].apply(lambda x: x[timeframe]))
        
        # Extract employment projection and salary data
        base_employment = df["baseEmployment"].values[0]
        projected_employment = [df[f"{timeframe}Projected"].values[0] for timeframe in timeframes]
        min_salary = df["min_salary"].values[0]
        avg_salary = df["average_salary"].values[0]
        max_salary = df["max_salary"].values[0]
        
        # Employment x-axis values (base and projections)
        x_labels = ["Base", "1 Year", "3 Years", "5 Years", "10 Years"]
        x_values = range(len(x_labels))

        # Create dual-axis plot
        plt.figure(figsize=(12, 7))

        fig, ax1 = plt.subplots(figsize=(10, 6))
        
        # Plot employment on the first y-axis
        ax1.plot(x_values, [base_employment] + projected_employment, marker='o', color='tab:blue', label="Employment", linestyle='-')
        ax1.set_xlabel('Years')
        ax1.set_ylabel('Employment', color='tab:blue')
        ax1.tick_params(axis='y', labelcolor='tab:blue')
        ax1.set_xticks(x_values)
        ax1.set_xticklabels(x_labels)
        
        # Add employment annotations
        for i, val in enumerate([base_employment] + projected_employment):
            ax1.annotate(f'{val}', (x_values[i], val), textcoords="offset points", xytext=(0,5), ha='center', fontsize=10)

        # Create a second y-axis for salary
        ax2 = ax1.twinx()
        
        # Plot salary data on the second y-axis
        ax2.plot([0, 1, 2, 3, 4], [min_salary] * 5, 'r--', label="Min Salary", color='tab:red')
        ax2.plot([0, 1, 2, 3, 4], [avg_salary] * 5, 'g--', label="Avg Salary", color='tab:green')
        ax2.plot([0, 1, 2, 3, 4], [max_salary] * 5, 'b--', label="Max Salary", color='tab:orange')
        
        # Set the labels and ticks for the second y-axis
        ax2.set_ylabel('Salary ($)', color='tab:red')
        ax2.tick_params(axis='y', labelcolor='tab:red')

        # Add a legend for both plots
        ax1.legend(loc='upper left')
        ax2.legend(loc='upper right')

        # Add title for the plot
        plt.title(f'Employment Growth and Salary Range for {df["name"].values[0]} in Region {df["region"].values[0]["name"]}')

        # Show plot
        plt.tight_layout()
        plt.show()
# Feeder jobs data, Next steps jobs data

# Valid values: "soc", "onet", "lotocc", "lotspecocc", "skill"
dimension = "lotspecocc"
# CTO -- 23141014 -- in the USA
occupation_id = "23141014"

# plot_feeder_job(url, dimension, occupation_id)
# plot_nextstep_jobs(url, dimension, occupation_id)
#########

# plot_occupation_data()

# plot_broad_comp_data()

# plot_title_comp_data()

plot_combined_title_occupation_data()
```
